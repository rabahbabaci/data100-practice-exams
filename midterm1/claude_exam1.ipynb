## Q1 — Café DataFrame [7.5 Pts]
````
 item category price rating available
latte Latte drink 4.50 4 True
muffin Blueberry Muffin food 3.25 5 True
tea Green Tea drink 3.00 3 False
sand Turkey Sandwich food 7.50 4 True
cake Chocolate Cake food 5.00 2 False
````
> ⚠️ The index is a **string label** (`'latte'`, `'muffin'`, etc.) — NOT the default integer index.

For each snippet: write the returned **integer (0–8)**, or **`NA`** if it errors or returns something else.


### (a) [1.5 Pts]
```python
menu['rating'][1]
```
**Answer:** ___

> 💡 **Solution:** `NA` — `menu['rating']` has a string index. `[1]` looks for label `1`, which doesn't exist → KeyError. Use `.iloc[1]` to get `5` by position.


### (b) [1.5 Pts]
```python
menu.iloc[2, 3]
```
**Answer:** ___

> 💡 **Solution:** `3` — `.iloc` uses integer positions. Row 2 = `tea`, column 3 = `rating` = `3`.


### (c) [1.5 Pts]
```python
menu.loc['sand', 'rating']
```
**Answer:** ___

> 💡 **Solution:** `4` — `.loc` uses labels. `'sand'` = Turkey Sandwich, `'rating'` = `4`.


**Answer:** ___

> 💡 **Solution:** `4` — Available rows: `latte, muffin, sand`. Ratings: `[4, 5, 4]`. `.iloc[0]` = `4`.
