## Q6. Sampling Logic (6 pts)
In the Bay Area traveler pool, 40% of passengers plan to take rideshare to the airport. Let \(\hat{p}\) be the sample proportion of rideshare users.

1. (2 pts) You draw a simple random sample of 100 travelers. What are \(\mathbb{E}[\hat{p}]\) and the standard error of \(\hat{p}\)? Show formulas + numbers.
2. (2 pts) Instead, you intercept only riders already waiting in the rideshare pickup curb. Is \(\hat{p}\) unbiased in this plan? Identify the bias type and explain.
3. (2 pts) Name two concrete ways to reduce **chance error** (not bias) for this estimate.

<details><summary>Solution</summary>
1. \(\mathbb{E}[\hat{p}] = p = 0.40\). SE for a proportion from SRS: \(\sqrt{p(1-p)/n} = \sqrt{0.4 \cdot 0.6 / 100} = \sqrt{0.24/100} = 0.049\).
2. No—sampling only at the rideshare curb over-represents rideshare users. That induces **selection bias** (convenience sample). \(\hat{p}\) will overestimate the true population proportion.
3. Increase sample size (e.g., survey more travelers) and/or use stratified sampling across time/location to keep randomness while shrinking variance. Another lever is repeated independent samples and averaging.
</details>

## Q7. Quick Probability Check (2 pts)
From the same traveler population (40% rideshare), you sample two travelers **with replacement**. What is the probability both selected travelers plan to use rideshare?

<details><summary>Solution</summary>
Each draw has probability 0.40 of being a rideshare user. With replacement the draws are independent, so \(0.40^2 = 0.16\).
</details>
