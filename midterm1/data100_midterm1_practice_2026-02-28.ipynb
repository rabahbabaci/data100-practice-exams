## Q2 — Grouped Orders [4.5 Pts]
Using the same `menu` DataFrame, each snippet returns a DataFrame or Series. Write the **number of rows** in the result (or `NA`).

```
 item category price rating available
latte Latte drink 4.50 4 True
muffin Blueberry Muffin food 3.25 5 True
tea Green Tea drink 3.00 3 False
sand Turkey Sandwich food 7.50 4 True
cake Chocolate Cake food 5.00 2 False
```

### (a) [1.5 Pts]
```python
menu.groupby('category')[['price']].agg('mean')
```
Answer: ___ (Solution: `2` rows for drink & food.)

### (b) [1.5 Pts]
```python
menu.groupby('category').head(1)
```
Answer: ___ (Solution: `2` rows — first original row of each group.)

### (c) [1.5 Pts]
```python
menu.groupby('category').filter(lambda sf: sf['rating'].mean() > 3)
```
Answer: ___ (Solution: `5` rows — `.filter` returns original rows for groups whose mean rating > 3.)

## Q6 — KDE & Distributions [4 Pts]
Points: {2, 6, 10}. Boxcar kernel with α = 4 (half-width 2).

### (a)
At x = 5: only point 6 contributes (|5−6|≤2). Contribution = 1/α = 1/4. KDE: (1/3)*(1/4) = **1/12 ≈ 0.0833**.

### (b)
- “Larger α always more accurate” → **False** (oversmoothing).
- “Area under KDE = 1” → **True**.

### (c)
Increasing α to 8 makes the curve **smoother and wider**.
